```python
import pandas as pd

# Creating a template for the Titanic Kaggle submission based on the user's specific requirements.
# PassengerId starts from 892 to 1309 (total 418 entries) based on standard Titanic test data.
passenger_ids = list(range(892, 1310))

# Creating a dummy 'Survived' column with 0s as a placeholder for the user to fill with their predictions.
df_submission = pd.DataFrame({
    'PassengerId': passenger_ids,
    'Survived': [0] * 418
})

# Saving to CSV as requested
file_path = 'titanic_submission.csv'
df_submission.to_csv(file_path, index=False)

print(f"File created with {len(df_submission)} rows and columns: {list(df_submission.columns)}")
```
